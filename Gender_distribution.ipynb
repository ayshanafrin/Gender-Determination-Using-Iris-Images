```python
import matplotlib.pyplot as plt

# Dataset counts
labels = ['Female', 'Male']
counts = [392, 312]

# Create bar chart
plt.figure(figsize=(6,4))
bars = plt.bar(labels, counts, color=['pink', 'lightblue'])

# Add numbers on top of bars
for bar in bars:
    yval = bar.get_height()
    plt.text(bar.get_x() + bar.get_width()/2, yval + 5, yval, ha='center', va='bottom', fontsize=12)

# Title and labels
plt.title('Gender Class Distribution', fontsize=14)
plt.ylabel('Number of Subjects', fontsize=12)
plt.ylim(0, max(counts)+50)

# Show the plot
plt.show()
```

```python
import os
import matplotlib.pyplot as plt

# Path to your dataset
dataset_path = '/content/drive/MyDrive/Mini_proj/Results'  # Change to your path

# List of folders
folders = ['Male_L', 'Male_R', 'Female_L', 'Female_R']

# Count images in each folder
counts = []
for folder in folders:
    folder_path = os.path.join(dataset_path, folder)
    if os.path.exists(folder_path):
        count = len([f for f in os.listdir(folder_path) if f.lower().endswith(('.jpg','.png','.jpeg'))])
    else:
        count = 0
    counts.append(count)

# Check if dataset is empty
total = sum(counts)
if total == 0:
    print("No images found in the dataset!")
else:
    percentages = [c/total*100 for c in counts]

    # ---------------- Bar Chart ----------------
    plt.figure(figsize=(12,5))
    bars = plt.bar(folders, counts, color=['lightblue','blue','pink','red'])

    # Add count and percentage on top
    for i, bar in enumerate(bars):
        yval = bar.get_height()
        plt.text(bar.get_x() + bar.get_width()/2, yval + 5, f'{counts[i]}\n({percentages[i]:.1f}%)',
                 ha='center', va='bottom', fontsize=10)

    plt.title('Gender Class Distribution (Bar Chart)', fontsize=14)
    plt.ylabel('Number of Images', fontsize=12)
    plt.ylim(0, max(counts)+50)
    plt.show()

    # ---------------- Pie Chart ----------------
    plt.figure(figsize=(6,6))
    colors = ['lightblue','blue','pink','red']
    plt.pie(counts, labels=folders, autopct='%1.1f%%', colors=colors, startangle=90)
    plt.title('Gender Class Distribution (Pie Chart)', fontsize=14)
    plt.show()
```
